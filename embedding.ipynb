```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### one hot Repre
one_hot_rep= [one_hot(words,voc_size)for words in sent]
one_hot_rep
# it is just giving the idea of that for examle word "of" is having 637 index "1" and rest other are zero
```

```python
##word embeding representation 
from pyexpat import model
from tensorflow.keras.layers import Embedding
#from tensorflow.kearas.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
##we need to to make same size of all the sentences
#maxmium decided as seen from data sentences, 
sent_len =8 

#padding will add zeros in the places where sent size is less
#if we give pre zeros 0 will be added pre or post
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_len)
embedded_docs
```

```python
##feature representation
##which means i want 10 dimension
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
